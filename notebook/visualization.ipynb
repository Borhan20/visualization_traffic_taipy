```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
event_data['event_date'] =   pd.to_datetime(event_data['event_time'])

time_series = event_data.groupby(pd.Grouper(key='event_date', freq='M')).size().reset_index(name='count')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_series['event_date'], time_series['count'],marker='o', linestyle='-', color='#1f77b4')
# Format x-axis dates
date_format = DateFormatter("%b %Y")
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()

# Add labels and title
plt.xlabel("Event Time", fontsize=12)
plt.ylabel("Count of Events", fontsize=12)
plt.title("Event Activity Over Time", fontsize=14, pad=20)
plt.grid(True, alpha=0.3)

# Show plot
plt.tight_layout()
plt.show()
```

```python
def map_site(referrer):
    if pd.isna(referrer):
        return 'other'
    
    referrer = str(referrer).lower()
    
    if 'linkedin' in referrer or 'lnkd.in' in referrer:
        return 'Linkedin'
    elif 'google' in referrer:
        return 'Google'
    elif 'instagram' in referrer:
        return 'Instagram'
    elif 'twitter' in referrer:
        return 'Twitter'
    else:
        return 'other'
```

```python
event_data['site_mapped']   = event_data['referrer'].apply(map_site)
print(event_data[['referrer', 'site_mapped']].head(10))
```

```python
# Group by month and site_mapped, then count events
time_series = event_data.groupby([pd.Grouper(key='event_date', freq='M'), 'site_mapped']).size().reset_index(name='count')

# Plot configuration
plt.figure(figsize=(14, 8))
date_format = DateFormatter("%b %Y")

# Plot each category separately
for site, data in time_series.groupby("site_mapped"):
    plt.plot(data['event_date'], data['count'], marker='o', linestyle='-', label=site)

# Formatting
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()
plt.title("Event Activity by Referral Source", fontsize=16, pad=20)
plt.xlabel("Event Time", fontsize=12)
plt.ylabel("Number of Events", fontsize=12)
plt.legend(title='Referral Source', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python

# Group and filter the data
os_counts = device_data.groupby('os_type').size().reset_index(name='count')
os_filtered = os_counts[os_counts['count'] > 100]
```

```python
# Create the plot
plt.figure(figsize=(12, 6))
bars = plt.bar(os_filtered['os_type'], os_filtered['count'], color='skyblue')

# Customize the plot
plt.title('OS Type Distribution (Count > 100)', pad=20, fontsize=14)
plt.xlabel('Operating System Type', fontsize=12)
plt.ylabel('Device Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:,}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
